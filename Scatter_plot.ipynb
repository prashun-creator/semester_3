###Plotting

####Using Matplotlib 

```python
plt.scatter(cars_data['City MPG'],cars_data['Hwy MPG'])
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
```

```python
cars_data['HP'].min()
```

```python
cars_data['HP'].max()
```

```python

colors=cars_data['HP']
plt.scatter(cars_data['City MPG'],cars_data['Hwy MPG'],c=colors,cmap='viridis')
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10, 6))   #increases figure size to see data points more clearly. 
colors=cars_data['HP']
plt.scatter(cars_data['City MPG'],cars_data['Hwy MPG'],c=colors,cmap='nipy_spectral')
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
plt.colorbar()  #required for right side bar which shows the HP columns.
plt.show()



```
